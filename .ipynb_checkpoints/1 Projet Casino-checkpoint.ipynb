## 3 façons de le faire, avec/sans True et la façon naïve
**Q : Que vaut mieux utiliser `random.randrange` ou `random.randint` ?**  
R :
https://stackoverflow.com/questions/3540431/difference-between-random-randint-vs-randrange

```python
# 1è façon avec True
from random import randrange
nb_ordi = randrange (0, 11, 1)
print ('Mon choix est = ', nb_ordi)
nb_coup = 0
while True :
	nb_user = int(input ("Entrez SVP votre nombre ? "))
	nb_coup += 1
	if nb_user > nb_ordi :
		print ('Votre nbre est trop grand')
	elif nb_user < nb_ordi :
		print ('Votre nbre est trop petit')
	else :
		print ("Bingo ! Vous avez gagné en {} coup(s) !".format(nb_coup))
		break
```

```python
# 2è façon sans True
from random import randrange
nb_ordi = randrange(0, 11, 1)
print("Mon choix = ", nb_ordi)
nb_user = -1
nb_coup = 0
while nb_ordi != nb_user :
	nb_user = int(input("Devine le nombre auquel je pense qui est compris entre 0 et 10\n"))
	if nb_user > nb_ordi :
		print ('Votre nbre est trop grand')
	elif nb_user < nb_ordi :
		print ('Votre nbre est trop petit')
	nb_coup += 1
print("Bingo ! Vous avez gagné en {} coup(s) !".format(nb_coup))
```

```python
# 3è façon : 
# Que pensez-vs de ce code ? 
# C'est une façon naïve de coder ! ça suit la logique du USER. Aussi, je me repete : ligne 7 et 14 [Entrez SVP votre nombre ? ] (dés que je me repete c pas bon).
from random import randrange
nb_ordi = randrange(0, 11, 1)
print("Mon choix est : ", nb_ordi)
nb_user = int(input("Entrez SVP votre nombre ? ")) 
nb_coup = 1
while (nb_ordi != nb_user) : 
	if(nb_user > nb_ordi) : 
		print("Votre nbre est trop grand") 
	else : 
		print("Votre nbre est trop petit")
	nb_user = int(input("Entrez SVP votre nombre ? "))
	nb_coup += 1
print("Bingo ! Vous avez gagné en {} coup(s) !".format(nb_coup))
```
